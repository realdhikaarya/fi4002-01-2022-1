```python
import numpy as np
import matplotlib.pyplot as plt
plt.style.use('seaborn-poster')
%matplotlib inline

n = 10
tbeg = 0
tend = 4
h = (tend-tbeg) / n

t1 = 2
y1 = 20

# Get A
A = np.zeros((n+1, n+1))
A[0, 0] = 1
A[n, n] = 1
for i in range(1, n):
    A[i, i-1] = 1
    A[i, i] = -2
    A[i, i+1] = 1

print(A)

# Get b
b = np.zeros(n+1)
b[0] = 0
b[1:-1] = -10*h**2
b[-1] = 0
print(b)

# solve the linear equations
y = np.linalg.solve(A, b)

t = np.linspace(tbeg, tend, n+1)

plt.figure(figsize=(10,8))
plt.plot(t, y)
plt.plot(t1, y1, 'ro')
plt.xlabel('time (s)')
plt.ylabel('altitude (m)')
plt.show()
```
